```python
def form_y(x, a=1,freq=1,phase=1):
    return a*np.sin(freq*x+phase)
def plot(x,y):
    plt.plot(x, y)
    plt.xlabel('Angle [rad]')
    plt.ylabel('sin(x)')
    plt.axis('tight')
    plt.show()
```

```python
r = 3 # сколько штук волн
x = np.linspace(-r*np.pi, r*np.pi, 10000)
print(x.shape)
```

```python
y = form_y(x)
plot(x, y)
```

```python
y2 = form_y(x, a = 23,freq=1, phase=1)
plot(x, y2)
```

```python
y3 = form_y(x, a = 23,freq=1, phase=1)
plot(x, y3)
```

```python
plot(x, y+y2+y3)
```

```python
y4 = form_y(x, a = -23,freq=3, phase=1)
plot(x, y4)
```

```python
plot(x, y+y2+y4)
```

```python
# fft transform for sinus
sp = np.fft.fft(y+y2+y4)
```

```python
freq = np.fft.fftfreq(y.shape[-1])
plt.plot(freq, sp.real, freq, sp.imag)
plt.show()
```
